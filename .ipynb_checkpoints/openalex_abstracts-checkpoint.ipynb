```python
import requests 
import numpy as np
```

```python
query_string1 = "environmental flows"
query_string2 = "habitat analysis"
query_string3 = "watershed model"
query_string4 = "tennant method"
query_string5 = "global environmental flow information system"
query_string6 = "ecohydrology"
search = requests.get(
    f"https://api.openalex.org/works?"
    f"filter="
    f"title.search:{query_string1}"
#     f"abstract.search:{query_string1}"
#     f"fulltext.search:{query_string1}"
    f"|"
    f"{query_string2}"
    f"|"
    f"{query_string3}"
    f"|"
    f"{query_string4}"
    f"|"
    f"{query_string5}"
    f"|"
    f"{query_string6}"
    f"|"
    f"&sort=relevance_score:desc&page=1&per-page=20"
).json()

resultlist = []

for x in range(10):
    result = {}
    for y in search['results'][x].keys():
        result[y] = search['results'][x][y]
    resultlist.append(result)

x = [i['title'] for i in resultlist]
y = [i['authorships'][0]['author']['display_name'] for i in resultlist]
z = [i['doi'] for i in resultlist]
zz = [i['relevance_score'] for i in resultlist]

for idx, i in enumerate(x):
    print(i)
    print(y[idx])
    print(z[idx])
    print(zz[idx])
    print('\n')
```
